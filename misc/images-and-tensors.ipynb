```python
import fastbook
fastbook.setup_book()
from fastai.vision.all import *
from fastbook import *
```

```python
path = Path('images')
path
```

```python
im = PILImage.create(path/'pup.jpg')
```

```python
type(im)
```

```python
im.show()
```

```python
im_tensor = tensor(im)
```

```python
type(im_tensor)
```

```python
im_tensor.shape
```

```python
im_tensor[220][220]
```

```python
im_tensor = im_tensor.permute(2, 0, 1)
im_tensor.shape
```

```python
im_tensor[0][220][220]
```

```python
import torchvision.transforms as T
```

```python
torchvis_transform = T.ToTensor()
```

```python
img_torchvis = torchvis_transform(im)
```

```python
img_torchvis.shape
```

```python
img_torchvis[0][220][220]
```

```python
type(img_torchvis)
```

```python
doc(ToTensor)
```

```python
transform = ToTensor()
```

```python
img_tensor = transform(im)
```

```python
type(img_tensor)
```

```python
img_tensor.shape
```

```python
img_tensor[0][220][220]
```

```python
img_tensor = img_tensor / 255
```

```python
img_tensor[0][220][220]
```

```python
transform_to_im = T.ToPILImage()
```

```python
image_from_tensor = transform_to_im(im_tensor)
```

```python
image_from_tensor
```

```python
PILImage(image_from_tensor).show()
```

```python
resizeTo = 224,224
```

```python
image_from_tensor = image_from_tensor.resize(resizeTo)
```
