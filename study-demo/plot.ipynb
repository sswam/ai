```python
#|export
import matplotlib.pyplot as plt
import numpy as np
```

```python
y = [4,5,6,7,8,9]
x = range(len(y))
plt.bar(x, y)
```

```python
#|export
def bars(y):
    x = range(len(y))
    plt.bar(x, y)
    plt.show()
```

```python
bars([4,5,6,7,8,9])
```

```python
#|export
def lines(x=None, y=None, y0=None, eq=False):
    plt.plot(x, y)
    if eq:
        plt.axis('equal')
    if y0 is not None:
        plt.ylim(bottom=y0)
    plt.show()
```

```python
lines([1,2,1.5], [9,8,7], y0=0, eq=True)
```

```python
x = np.arange(-4, 4, 0.01)
y = np.sin(x**3)
# lines(x, y)
plt.plot(x, y)
```

```python
a = np.arange(0, 60, 0.01)
x = (a + np.sin(a**2)) * np.sin(a)
y = (a + np.sin(a**2)) * np.cos(a)
lines(x, y, eq=True)
```
